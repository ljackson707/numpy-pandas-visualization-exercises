#### Types of data:

- Continuous Data:- Numeric data with possibly infinite resolution
    - Height
    - Weight
    - total_bill and tip
    - temperature
- Categorical - Distinct categories
    - weekdays
    - Gender
    - 'smoker'
    - letter grades (A, B, C..)

- Discrete - Distinct numeric categories 
     - party_size - numeric but discete
     - number of customer complaints
     - number of flaws or defects.

### Relational Plots 

```python
# Explore relationship between total_bill and tip
sns.set() # Seaborn default style
sns.relplot(data = tips, x = 'total_bill', y = 'tip')

plt.title("Total Bill vs Tip Ammount")
# returns a FacetGrid (which means a figure level object)
```

```python
# viz if dist is diffrent forsmoker vs non-smoker
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue='smoker')
```

```python
# by time of day
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue='day')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue='time')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', style='time') 
#gives diffrent style of marker insted of color 
```

```python
# use arggument called col
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue = 'time', col = 'day')
           
# can add , col_wrap = 2) to limit size of overall box
```

#### Key takeaways?
- Tip ammount generally increses with total_bill
- Waiter works in evenings on Sun and Sat
- Waiter works during lunchtime on Thursday but not on Sun or Sat

### Categorial Plots

#### Boxplots

```python
sns.boxplot(data = tips, x = "tip")
plt.grid(True, ls = "--")
plt.xticks(np.arange(0,10,1));
```

```python
# line in middle of box = median
# left side of line = 1 and 2 quartile
# right side of the line = 3 quatile
# interquartile range = (q3 - q1) 
# dots on left = min outliers 
# dots on right = max outliers 
# t bracket left and right = min and max 
```

```python
tips.tip.describe()
```

```python
plt.figure(figsize = (10,8))

plt.subplot(221)
sns.boxplot(y = 'tip', data = tips)

plt.subplot(222)
sns.boxplot(x = 'tip', data = tips)

plt.subplot(223)
sns.boxplot(y = 'tip', x = 'size', data = tips)

plt.subplot(224)
sns.boxplot(x = 'tip', y = 'time', data = tips, hue = 'smoker',
            order = ['Lunch', 'Dinner'], hue_order = ['No', 'Yes']) # changes the order of y axis
```

#### Heatmap

```python
data = pd.crosstab(tips.time, tips['size'])
```

```python
sns.heatmap(data, cmap = plt.cm.Greens, annot = True, fmt = 'd')
```

```python
data2 = pd.crosstab(tips.time, tips.smoker)
```

```python
sns.heatmap(data2, annot = True, fmt = 'd')
```
