```python
#1) Copy the code from the lesson to create a dataframe full of student grades.
np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
df
```

```python
df.shape
```

```python
df.info()
```

```python
#A) Create a column named passing_english that indicates whether each student has a passing grade in english.
new_df = df.assign(passing_english = df.english >= 70)
new_df
```

```python
#B)Sort the english grades by the passing_english column. How are duplicates handled?
new_df.sort_values(by = 'passing_english') # drops the duplicates, we need to use a keep all function to keep this from happening
```

```python
new_df['passing_english'].sum()
```

```python
#C) Sort the english grades first by passing_english and then by student name. 
#All the students that are failing english should be first, 
#and within the students that are failing english they should be ordered alphabetically. 
#The same should be true for the students passing english. 
#(Hint: you can pass a list to the .sort_values method)

new_df.sort_values(by=['passing_english', 'name'], ascending = [True, True])
```

```python
(new_df['passing_english'] == False).sum()
```

```python
#D) Sort the english grades first by passing_english, and then by the actual english grade, 
#similar to how we did in the last step.
new_df.sort_values(by=['passing_english', 'english'], ascending = [False, False])
```

```python
#E) Calculate each students overall grade and add it as a column on the dataframe. 
#The overall grade is the average of the math, english, and reading grades.
new_df = df.assign(overall_grade = (df.math+df.english+df.reading)/3)
new_df.round(0)
```

```python
# This method also works
new_df['overall_average'] = round(new_df.loc[:, 'math': 'reading'].sum(axis=1)/3).astype(int)
new_df
```

```python
#2) Load the mpg dataset. 
#Read the documentation for the dataset and use it for the following questions:
mpg = data('mpg')
mpg
```

```python
#a) How many rows and columns are there?
mpg.shape
```

```python
#b) What are the data types of each column?
mpg.info()
```

```python
#c) Summarize the dataframe with .info and .describe
mpg.describe()
```

```python
#d) Rename the cty column to city.
new_mpg = mpg.rename(columns={'cty': 'city'})
new_mpg.columns
```

```python
#e) Rename the hwy column to highway.
new_mpg = mpg.rename(columns={'hwy': 'highway'})
new_mpg.columns
```

```python
#f) Do any cars have better city mileage than highway mileage?
new_mpg = mpg.sort_values(by=['city', 'highway'], ascending = [False, False])
new_mpg
```

```python
#g) Create a column named mileage_difference 
#this column should contain the difference between highway and city mileage for each car.
new_mpg = (mpg.assign(mileage_difference = (mpg.highway - mpg.city))
.sort_values(by = 'mileage_difference', ascending = False))
new_mpg 
```

```python
#h) Which car (or cars) has the highest mileage difference?
new_mpg.nlargest(1, 'mileage_difference', keep='all')
```

```python
#i) Which compact class car has the lowest highway mileage? The best?
new_mpg_low = (new_mpg[new_mpg['class'] == 'compact']
.sort_values(by = 'highway', ascending = False)
.tail(1))
new_mpg_low
```

```python
# The best
new_mpg_high = (new_mpg[new_mpg['class'] == 'compact']
.sort_values(by = 'highway', ascending = False)
.head(1))
new_mpg_high
```

```python
#j) Create a column named average_mileage that is the mean of the city and highway mileage.
new_mpg = (mpg.assign(average_mileage = (mpg.highway + mpg.city)/2)
.sort_values(by = 'average_mileage', ascending = False))
new_mpg
```

```python
#k) Which dodge car has the best average mileage? The worst?

# The best
new_mpg_h = (new_mpg[new_mpg['manufacturer'] == 'dodge']
.sort_values(by = 'average_mileage', ascending = False)
.head(1))
new_mpg_h
```

```python
# The Worst
new_mpg_l = (new_mpg[new_mpg['manufacturer'] == 'dodge']
.sort_values(by = 'average_mileage', ascending = False)
.tail(4))
new_mpg_l
```

```python
#d) What is the the weight of the fastest animal?
weight_of_fast = mammals.sort_values(by = 'speed', ascending = False).head(1)
weight_of_fast
```

```python
#e) What is the overal percentage of specials?
mammals.specials.value_counts(normalize=True).mul(100).astype(str)+'%'
```

```python
#f) How many animals are hoppers that are above the median speed? What percentage is this?
med_speed= (sum(mammals.speed)/107)
med_speed

hoppers_over_mspeed = (mammals.assign(percent_hop_over_med_speed = (mammals.hoppers == 1) & (mammals.speed > med_speed))
.sort_values(by = ['hoppers','percent_hop_over_med_speed'], ascending = [False, False] ))
hoppers_over_mspeed


hoppers_over_mspeed.percent_hop_over_med_speed.value_counts(normalize=True).mul(100).astype(str)+'%'
```
